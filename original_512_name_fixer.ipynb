## Fixing Generated File Names as well \<sigh\>

```python
genned_path = './processed/'
```

```python
df = pd.DataFrame(os.listdir(genned_path))
df.columns = ['file_names']
df.head()
```

```python
df['file_names'].unique()
```

```python
files_to_rename_genned = df[df['file_names'].apply(lambda x: len(x.split('_')[:-2]) > 1)]
#df[df['file_names'].apply(lambda x: len(x.split('_')[:-2]) > 1)]
files_to_rename_genned
```

```python
files_to_rename_genned['id'] = list(files_to_rename_genned.index);
files_to_rename_genned = files_to_rename_genned.reset_index(drop = True)
files_to_rename_genned
```

```python
#files_to_rename_genned
new_names_genned = []
for i in range(len(files_to_rename_genned)):
    #print(files_to_rename_genned.iloc[i, 0].split('_')[:-2])
    #print('.'.join(files_to_rename_genned.iloc[i, 0].split('_')[:-2]))
    #print ('_'.join(files_to_rename_genned.iloc[i, 0].split('_')[1:]), '\n')
    first_temp = '+'.join(files_to_rename_genned.iloc[i, 0].split('_')[:-2])
    second_temp = '_'.join(files_to_rename_genned.iloc[i, 0].split('_')[1:])
    new_name = first_temp + '.' + second_temp
    #print (new_name)
    new_names_genned.append(new_name)
    
new_names_genned
```

```python
files_to_rename_genned ['new_filename'] = new_names_genned;
files_to_rename_genned.head()
```

```python

for i in range(len(files_to_rename_genned)):
    filepath = genned_path + files_to_rename_genned['file_names'][i];
    filename_to = genned_path + files_to_rename_genned['new_filename'][i];
    os.rename(src = filepath, dst = filename_to)
```
