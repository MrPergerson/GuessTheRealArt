```python
#piece_title = df['Artist_name'].apply()

def work_title_extractor (df_fn_lst, df_artist_names):
    index_lst = range(len(df_fn_lst));
    #temp = [i for i in df_fn_lst if i.split('-') not in df_artist_names]
    temp = []
    for i in index_lst:
        extracted = [j for j in df_fn_lst[i].split('-') if j not in df_artist_names[i]]
        #extracted.append(j for j in df_fn_lst[i].split('-') if j not in df_artist_names[i])
        temp.append(extracted)
    return pd.Series(temp)
```
